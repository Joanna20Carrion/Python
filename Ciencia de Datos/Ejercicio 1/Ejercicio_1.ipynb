```python
# Importar librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV
url = "C:/Users/joann/Desktop/Codigos/Trabajo_Final_Joanna_Carrión/billboard_24years_lyrics_spotify.csv"
df = pd.read_csv(url)

# Mostrar las primeras filas del dataframe
df.head()
```

```python
# 1. Tratamiento de Valores Nulos
# Identificación de valores nulos
print("Valores nulos por columna:")
print(df.isnull().sum())

# Reemplazar valores nulos por 0
df = df.fillna(0)

# Verificación después del reemplazo de valores nulos
print("Valores nulos por columna después del reemplazo:")
print(df.isnull().sum())
```

```python
# 2. Tratamiento de Valores Duplicados
# Identificación de duplicados
print("Filas duplicadas antes de la eliminación:")
print(df.duplicated().sum())

df = df.drop_duplicates()

print("Filas duplicadas después de la eliminación:")
print(df.duplicated().sum())
```

```python
# 3. Normalización de Datos
# Identificación de variables pertinentes
# Suponiendo que las variables pertinentes son numéricas como 'popularity', 'duration_ms', etc.
numeric_columns = ['danceability','energy','loudness','speechiness','acousticness','instrumentalness','liveness','valence']

# Normalización Min-Max
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])

# Verificar las primeras filas después de la normalización
print(df.head())
```

```python
# 4. Identificación y Eliminación de Variables Redundantes
# Seleccionar solo las columnas numéricas para el cálculo de la matriz de correlación
numeric_df = df.select_dtypes(include=[np.number])

# Identificación de variables redundantes mediante correlación
corr_matrix = numeric_df.corr().abs()
high_corr_var = np.where(corr_matrix > 0.8)
high_corr_var = [(corr_matrix.index[x], corr_matrix.columns[y]) for x, y in zip(*high_corr_var) if x != y and x < y]
print('Se eliminan las siguiente variables altamente correlacionales: ',high_corr_var)

# Se elimina una de cada par de variables altamente correlacionadas
for var in high_corr_var:
    if var[1] in df.columns:
        df = df.drop(var[1], axis=1)

# Verificar el DataFrame final
df.info()

# Guardar el DataFrame preprocesado en un nuevo archivo CSV
df.to_csv('billboard_24years_lyrics_spotify_preprocessed.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Suprimir advertencias futuras
warnings.simplefilter(action='ignore', category=FutureWarning)

# Cargar el dataset preprocesado
df = pd.read_csv('billboard_24years_lyrics_spotify_preprocessed.csv')

# 1. Distribución de la Popularidad de las Canciones por Año
plt.figure(figsize=(14, 7))
sns.histplot(data=df, x='year', bins=30, kde=True)
plt.title('Distribución de la Popularidad de las Canciones por Año')
plt.xlabel('Año')
plt.ylabel('Cantidad de Canciones')
plt.show()

# Descripción e Interpretación
# La visualización muestra la distribución del número de canciones populares por año en el período de 24 años.
# La gráfica revela cómo la cantidad de canciones populares ha cambiado a lo largo de los años.
# Podemos observar posibles picos en ciertos años, lo cual podría indicar periodos en los que se lanzaron muchas canciones populares.
```

```python
# 2. Relación entre Danceability y Acousticness
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='danceability', y='acousticness', hue='year', palette='viridis')
plt.title('Relación entre Danceability y Acousticness')
plt.xlabel('Danceability')
plt.ylabel('Acousticness')
plt.legend(title='Año', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Descripción e Interpretación
# Este scatter plot muestra la relación entre la capacidad de baile de las canciones y su cualidad acústica.
# Los puntos están coloreados según el año, lo que permite ver si hay alguna tendencia temporal en estos atributos.
# Se puede observar si las canciones con mayor danceability tienden a tener menor acousticness y viceversa.
```

```python
# 3. Boxplot de la Popularidad por Año
plt.figure(figsize=(14, 7))
sns.boxplot(data=df, x='year', y='danceability')
plt.title('Distribución de Danceability por Año')
plt.xlabel('Año')
plt.ylabel('Danceability')
plt.show()

# Descripción e Interpretación
# El boxplot muestra la distribución de la danceability de las canciones por año.
# Permite identificar la mediana, cuartiles y posibles outliers en la capacidad de baile de las canciones a lo largo de los años.
# Esto ayuda a entender cómo ha cambiado la danceability de las canciones populares con el tiempo.
```

```python
# 4. Mapeo de Bandas/Cantantes más Frecuentes en el Ranking
genre_count = df['band_singer'].value_counts().head(10)
plt.figure(figsize=(12, 8))
sns.barplot(x=genre_count, y=genre_count.index, palette='muted')
plt.title('Top 10 Bandas/Cantantes más Frecuentes en el Ranking')
plt.xlabel('Número de Canciones en el Ranking')
plt.ylabel('Bandas/Cantantes')
plt.show()

# Descripción e Interpretación
# La visualización muestra un gráfico de barras horizontal que destaca las 10 bandas o cantantes más frecuentes en el ranking de canciones.
# En el eje y, se enumeran las bandas o cantantes, mientras que en el eje x se muestra el número de canciones de cada banda o cantante que han aparecido en el ranking.
# Esta gráfica proporciona una visión clara de cuáles son los artistas más dominantes en el ranking de canciones populares.
# Al identificar las bandas o cantantes con más canciones en el ranking, podemos deducir cuáles han tenido un impacto significativo y constante en la música durante el período estudiado.
# Esto puede ser indicativo de la popularidad y éxito continuo de estos artistas en la industria musical.
# Además, el gráfico puede ayudar a identificar posibles tendencias de popularidad entre diferentes géneros musicales, dependiendo de los artistas que aparezcan con mayor frecuencia.
```

```python
# 5. Análisis de la Distribución de Atributos de Canciones
plt.figure(figsize=(14, 10))
attributes = ['danceability', 'acousticness', 'speechiness', 'instrumentalness', 'liveness']
for i, attr in enumerate(attributes, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[attr], bins=30, kde=True, color='blue')
    plt.title(f'Distribución de {attr}')
    plt.xlabel(attr)
    plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()

# Descripción e Interpretación
# La visualización consiste en una serie de histogramas que muestran la distribución de cinco atributos musicales de las canciones: danceability, acousticness, speechiness, instrumentalness, y liveness.
# Cada subgráfico representa la distribución de uno de estos atributos, utilizando histogramas con densidad de kernel (KDE) superpuesta para proporcionar una visualización más suave de la distribución.
# Danceability: La distribución del atributo danceability indica qué tan bailables son las canciones. La forma del histograma revela si hay un rango específico de valores que son más comunes, lo que puede ayudar a entender qué tan variadas son las canciones en términos de su capacidad para ser bailadas.
# Acousticness: Este atributo mide la calidad acústica de las canciones. Un histograma con una alta concentración de valores cercanos a cero puede sugerir que la mayoría de las canciones en el dataset son eléctricas o no acústicas.
# Speechiness: Mide la presencia de características de habla en las canciones. Si la mayoría de los valores están cerca de cero, indica que las canciones suelen tener menos características de habla, como en los géneros musicales instrumentales.
# Instrumentalness: Este atributo evalúa la probabilidad de que una canción sea instrumental. Un histograma con valores bajos puede sugerir que pocas canciones en el dataset son completamente instrumentales.
# Liveness: Refleja la presencia de una audiencia en la grabación. Un rango amplio de valores sugiere diversidad en las grabaciones, desde estudios en solitario hasta actuaciones en vivo.
```

```python
# 6. Heatmap de Correlación entre Atributos Numéricos
plt.figure(figsize=(12, 8))
corr_matrix = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Mapa de Calor de la Correlación entre Atributos Numéricos')
plt.show()

# Descripción e Interpretación
# El mapa de calor muestra la matriz de correlación entre los atributos numéricos del dataset. Cada celda del mapa representa el coeficiente de correlación de Pearson entre dos atributos numéricos, que puede variar de -1 (correlación negativa perfecta) a 1 (correlación positiva perfecta).
# Las celdas están coloreadas de acuerdo con la intensidad de la correlación, utilizando un esquema de colores coolwarm, con tonos fríos que indican correlaciones negativas y tonos cálidos que indican correlaciones positivas. Las anotaciones en cada celda muestran el valor exacto de la correlación.
# Las celdas con valores cercanos a 1 indican una fuerte correlación positiva entre los atributos. Por ejemplo, si danceability y energy tienen una alta correlación positiva, significa que las canciones más bailables tienden a tener también más energía.
# Las celdas con valores cercanos a -1 indican una fuerte correlación negativa. Si acousticness y instrumentalness tienen una alta correlación negativa, esto puede sugerir que las canciones más acústicas tienden a ser menos instrumentales.
# Sin Correlación Significativa: Valores cercanos a 0 indican que no hay una correlación significativa entre los atributos. Por ejemplo, si liveness no muestra una correlación notable con otros atributos, esto sugiere que la presencia de una audiencia en la grabación no está relacionada con las características musicales medidas.
```
