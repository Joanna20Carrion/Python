```python
# Importar librerÃ­as necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV
url = "C:/Users/joann/Desktop/Codigos/Trabajo_Final_Joanna_CarriÃ³n/billboard_24years_lyrics_spotify.csv"
df = pd.read_csv(url)

# Mostrar las primeras filas del dataframe
df.head()
```

```python
# 1. Tratamiento de Valores Nulos
# IdentificaciÃ³n de valores nulos
print("Valores nulos por columna:")
print(df.isnull().sum())

# Reemplazar valores nulos por 0
df = df.fillna(0)

# VerificaciÃ³n despuÃ©s del reemplazo de valores nulos
print("Valores nulos por columna despuÃ©s del reemplazo:")
print(df.isnull().sum())
```

```python
# 2. Tratamiento de Valores Duplicados
# IdentificaciÃ³n de duplicados
print("Filas duplicadas antes de la eliminaciÃ³n:")
print(df.duplicated().sum())

df = df.drop_duplicates()

print("Filas duplicadas despuÃ©s de la eliminaciÃ³n:")
print(df.duplicated().sum())
```

```python
# 3. NormalizaciÃ³n de Datos
# IdentificaciÃ³n de variables pertinentes
# Suponiendo que las variables pertinentes son numÃ©ricas como 'popularity', 'duration_ms', etc.
numeric_columns = ['danceability','energy','loudness','speechiness','acousticness','instrumentalness','liveness','valence']

# NormalizaciÃ³n Min-Max
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])

# Verificar las primeras filas despuÃ©s de la normalizaciÃ³n
print(df.head())
```

```python
# 4. IdentificaciÃ³n y EliminaciÃ³n de Variables Redundantes
# Seleccionar solo las columnas numÃ©ricas para el cÃ¡lculo de la matriz de correlaciÃ³n
numeric_df = df.select_dtypes(include=[np.number])

# IdentificaciÃ³n de variables redundantes mediante correlaciÃ³n
corr_matrix = numeric_df.corr().abs()
high_corr_var = np.where(corr_matrix > 0.8)
high_corr_var = [(corr_matrix.index[x], corr_matrix.columns[y]) for x, y in zip(*high_corr_var) if x != y and x < y]
print('Se eliminan las siguiente variables altamente correlacionales: ',high_corr_var)

# Se elimina una de cada par de variables altamente correlacionadas
for var in high_corr_var:
    if var[1] in df.columns:
        df = df.drop(var[1], axis=1)

# Verificar el DataFrame final
df.info()

# Guardar el DataFrame preprocesado en un nuevo archivo CSV
df.to_csv('billboard_24years_lyrics_spotify_preprocessed.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Suprimir advertencias futuras
warnings.simplefilter(action='ignore', category=FutureWarning)

# Cargar el dataset preprocesado
df = pd.read_csv('billboard_24years_lyrics_spotify_preprocessed.csv')

# 1. DistribuciÃ³n de la Popularidad de las Canciones por AÃ±o
plt.figure(figsize=(14, 7))
sns.histplot(data=df, x='year', bins=30, kde=True)
plt.title('DistribuciÃ³n de la Popularidad de las Canciones por AÃ±o')
plt.xlabel('AÃ±o')
plt.ylabel('Cantidad de Canciones')
plt.show()

# DescripciÃ³n e InterpretaciÃ³n
# La visualizaciÃ³n muestra la distribuciÃ³n del nÃºmero de canciones populares por aÃ±o en el perÃ­odo de 24 aÃ±os.
# La grÃ¡fica revela cÃ³mo la cantidad de canciones populares ha cambiado a lo largo de los aÃ±os.
# Podemos observar posibles picos en ciertos aÃ±os, lo cual podrÃ­a indicar periodos en los que se lanzaron muchas canciones populares.
```

```python
# 2. RelaciÃ³n entre Danceability y Acousticness
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='danceability', y='acousticness', hue='year', palette='viridis')
plt.title('RelaciÃ³n entre Danceability y Acousticness')
plt.xlabel('Danceability')
plt.ylabel('Acousticness')
plt.legend(title='AÃ±o', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# DescripciÃ³n e InterpretaciÃ³n
# Este scatter plot muestra la relaciÃ³n entre la capacidad de baile de las canciones y su cualidad acÃºstica.
# Los puntos estÃ¡n coloreados segÃºn el aÃ±o, lo que permite ver si hay alguna tendencia temporal en estos atributos.
# Se puede observar si las canciones con mayor danceability tienden a tener menor acousticness y viceversa.
```

```python
# 3. Boxplot de la Popularidad por AÃ±o
plt.figure(figsize=(14, 7))
sns.boxplot(data=df, x='year', y='danceability')
plt.title('DistribuciÃ³n de Danceability por AÃ±o')
plt.xlabel('AÃ±o')
plt.ylabel('Danceability')
plt.show()

# DescripciÃ³n e InterpretaciÃ³n
# El boxplot muestra la distribuciÃ³n de la danceability de las canciones por aÃ±o.
# Permite identificar la mediana, cuartiles y posibles outliers en la capacidad de baile de las canciones a lo largo de los aÃ±os.
# Esto ayuda a entender cÃ³mo ha cambiado la danceability de las canciones populares con el tiempo.
```

```python
# 4. Mapeo de Bandas/Cantantes mÃ¡s Frecuentes en el Ranking
genre_count = df['band_singer'].value_counts().head(10)
plt.figure(figsize=(12, 8))
sns.barplot(x=genre_count, y=genre_count.index, palette='muted')
plt.title('Top 10 Bandas/Cantantes mÃ¡s Frecuentes en el Ranking')
plt.xlabel('NÃºmero de Canciones en el Ranking')
plt.ylabel('Bandas/Cantantes')
plt.show()

# DescripciÃ³n e InterpretaciÃ³n
# La visualizaciÃ³n muestra un grÃ¡fico de barras horizontal que destaca las 10 bandas o cantantes mÃ¡s frecuentes en el ranking de canciones.
# En el eje y, se enumeran las bandas o cantantes, mientras que en el eje x se muestra el nÃºmero de canciones de cada banda o cantante que han aparecido en el ranking.
# Esta grÃ¡fica proporciona una visiÃ³n clara de cuÃ¡les son los artistas mÃ¡s dominantes en el ranking de canciones populares.
# Al identificar las bandas o cantantes con mÃ¡s canciones en el ranking, podemos deducir cuÃ¡les han tenido un impacto significativo y constante en la mÃºsica durante el perÃ­odo estudiado.
# Esto puede ser indicativo de la popularidad y Ã©xito continuo de estos artistas en la industria musical.
# AdemÃ¡s, el grÃ¡fico puede ayudar a identificar posibles tendencias de popularidad entre diferentes gÃ©neros musicales, dependiendo de los artistas que aparezcan con mayor frecuencia.
```

```python
# 5. AnÃ¡lisis de la DistribuciÃ³n de Atributos de Canciones
plt.figure(figsize=(14, 10))
attributes = ['danceability', 'acousticness', 'speechiness', 'instrumentalness', 'liveness']
for i, attr in enumerate(attributes, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[attr], bins=30, kde=True, color='blue')
    plt.title(f'DistribuciÃ³n de {attr}')
    plt.xlabel(attr)
    plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()

# DescripciÃ³n e InterpretaciÃ³n
# La visualizaciÃ³n consiste en una serie de histogramas que muestran la distribuciÃ³n de cinco atributos musicales de las canciones: danceability, acousticness, speechiness, instrumentalness, y liveness.
# Cada subgrÃ¡fico representa la distribuciÃ³n de uno de estos atributos, utilizando histogramas con densidad de kernel (KDE) superpuesta para proporcionar una visualizaciÃ³n mÃ¡s suave de la distribuciÃ³n.
# Danceability: La distribuciÃ³n del atributo danceability indica quÃ© tan bailables son las canciones. La forma del histograma revela si hay un rango especÃ­fico de valores que son mÃ¡s comunes, lo que puede ayudar a entender quÃ© tan variadas son las canciones en tÃ©rminos de su capacidad para ser bailadas.
# Acousticness: Este atributo mide la calidad acÃºstica de las canciones. Un histograma con una alta concentraciÃ³n de valores cercanos a cero puede sugerir que la mayorÃ­a de las canciones en el dataset son elÃ©ctricas o no acÃºsticas.
# Speechiness: Mide la presencia de caracterÃ­sticas de habla en las canciones. Si la mayorÃ­a de los valores estÃ¡n cerca de cero, indica que las canciones suelen tener menos caracterÃ­sticas de habla, como en los gÃ©neros musicales instrumentales.
# Instrumentalness: Este atributo evalÃºa la probabilidad de que una canciÃ³n sea instrumental. Un histograma con valores bajos puede sugerir que pocas canciones en el dataset son completamente instrumentales.
# Liveness: Refleja la presencia de una audiencia en la grabaciÃ³n. Un rango amplio de valores sugiere diversidad en las grabaciones, desde estudios en solitario hasta actuaciones en vivo.
```

```python
# 6. Heatmap de CorrelaciÃ³n entre Atributos NumÃ©ricos
plt.figure(figsize=(12, 8))
corr_matrix = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Mapa de Calor de la CorrelaciÃ³n entre Atributos NumÃ©ricos')
plt.show()

# DescripciÃ³n e InterpretaciÃ³n
# El mapa de calor muestra la matriz de correlaciÃ³n entre los atributos numÃ©ricos del dataset. Cada celda del mapa representa el coeficiente de correlaciÃ³n de Pearson entre dos atributos numÃ©ricos, que puede variar de -1 (correlaciÃ³n negativa perfecta) a 1 (correlaciÃ³n positiva perfecta).
# Las celdas estÃ¡n coloreadas de acuerdo con la intensidad de la correlaciÃ³n, utilizando un esquema de colores coolwarm, con tonos frÃ­os que indican correlaciones negativas y tonos cÃ¡lidos que indican correlaciones positivas. Las anotaciones en cada celda muestran el valor exacto de la correlaciÃ³n.
# Las celdas con valores cercanos a 1 indican una fuerte correlaciÃ³n positiva entre los atributos. Por ejemplo, si danceability y energy tienen una alta correlaciÃ³n positiva, significa que las canciones mÃ¡s bailables tienden a tener tambiÃ©n mÃ¡s energÃ­a.
# Las celdas con valores cercanos a -1 indican una fuerte correlaciÃ³n negativa. Si acousticness y instrumentalness tienen una alta correlaciÃ³n negativa, esto puede sugerir que las canciones mÃ¡s acÃºsticas tienden a ser menos instrumentales.
# Sin CorrelaciÃ³n Significativa: Valores cercanos a 0 indican que no hay una correlaciÃ³n significativa entre los atributos. Por ejemplo, si liveness no muestra una correlaciÃ³n notable con otros atributos, esto sugiere que la presencia de una audiencia en la grabaciÃ³n no estÃ¡ relacionada con las caracterÃ­sticas musicales medidas.
```
